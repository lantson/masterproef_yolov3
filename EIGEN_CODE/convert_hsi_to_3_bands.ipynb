```python
import spectral as spectral
import spectral.io.envi as envi
import os
import uuid

import numpy as np
```

```python
PATH_DATA = "data"
TRAIN_DIRECTORY = PATH_DATA + "\\" + "train"
TEST_DIRECTORY = PATH_DATA + "\\" + "test"

REDUCED_BAND_DIRECTORY1 = TRAIN_DIRECTORY + "_" + "3bands"
REDUCED_BAND_DIRECTORY2 = TEST_DIRECTORY + "_" + "3bands"
```

```python
if not os.path.exists(REDUCED_BAND_DIRECTORY1):
    !mkdir {REDUCED_BAND_DIRECTORY1}
    
if not os.path.exists(REDUCED_BAND_DIRECTORY2):
    !mkdir {REDUCED_BAND_DIRECTORY2}
```

```python
def load_hyperspectral_image_and_reduce_to_3_bands(filename):
    print(filename)
    img = envi.open(filename + ".hdr", filename + ".raw")
    img = np.array(img.load())
    img = img[:,:,1:4]
    print(img.shape)
    return img
```

```python
_THE_DIRECTORY = TRAIN_DIRECTORY
_SAVE_DIRECTORY = REDUCED_BAND_DIRECTORY1

for filename in os.listdir(_THE_DIRECTORY):
    if filename.endswith(".raw"):
        
        filename = filename[:-4]
        print(filename)
        img = load_hyperspectral_image_and_reduce_to_3_bands(_THE_DIRECTORY + "/" + filename)

        envi.save_image(_SAVE_DIRECTORY + "/" + filename + ".hdr", img, dtype=np.float32, ext=".raw")
```
