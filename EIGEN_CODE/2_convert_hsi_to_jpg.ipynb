```python
import spectral as spectral
import spectral.io.envi as envi
from spectral import save_rgb
import os
import uuid

import numpy as np
```

```python
PATH_DATA = "data"
SWIR = "SWIR"
NIR = "NIR"
SUBSETTED_DATA = "24bands"
PATH_DIRECTORY = PATH_DATA + "\\" + NIR + "\\" + SUBSETTED_DATA
JPG_DIRECTORY = PATH_DATA + "\\" + NIR + "\\" + "jpg"
```

```python
if not os.path.exists(JPG_DIRECTORY):
    !mkdir {JPG_DIRECTORY}
```

```python
def load_hyperspectral_image(filename):
    filename = filename[:-4]
    print(filename)
    img = envi.open(filename + ".hdr", filename + ".raw")
    img = np.array(img.load())
    return img
```

```python
for filename in os.listdir(PATH_DIRECTORY):
    if filename.endswith(".raw"):
        
        img = load_hyperspectral_image(PATH_DIRECTORY + "/" + filename)
        filename = filename[:-4] + ".jpg"
        save_rgb(filename= JPG_DIRECTORY + "\\" + filename, data=img, bands=(4,8,12))
        
        ## bands
        # NIR => (4,8,12) is vrij duidelijk
```
